```python
import pandas as pd

# the relative path of the stroke csv dataset is saved into the variable data for ease of use
data = '../data/processed/stroke_copy_processed.csv'

# The data is read and saved in a variable (stroke)
stroke_processed = pd.read_csv(data)

stroke_processed = pd.DataFrame(stroke_processed)


stroke_processed
```

```python
# Capture categorical columns (same method you use)
cat_cols = stroke_processed_copy.select_dtypes(include=['object', 'category']).columns.tolist()
print("Categorical columns detected:", cat_cols)

# Record unique values for each categorical column
uniques = {col: stroke_processed_copy[col].dropna().unique().tolist() for col in cat_cols}
for col, vals in uniques.items():
    print(f"\n{col} ({len(vals)}): {vals}")
    

# Perform one-hot encoding and keep a reference to the encoded columns
stroke_processed_ohe = pd.get_dummies(
    stroke_processed_copy,
    columns=cat_cols,
    dummy_na=False,
    dtype=np.uint8
)

# Build mapping: original column -> category value -> resulting one-hot column name
mapping_rows = []
for col in cat_cols:
    for val in uniques[col]:
        one_hot_col = f"{col}_{val}"
        mapping_rows.append({"original_col": col, "category_value": val, "one_hot_col": one_hot_col})

mapping_df = pd.DataFrame(mapping_rows)

# setting a 1-based index using an Index object (avoids static type issues)
mapping_df.index = pd.Index(range(1, len(mapping_df) + 1))
print("\nMapping (original column, category value -> one-hot column name):")
mapping_df.head(30)  # show all (adjust if many)

# save mapping to CSV for documentation:
# mapping_df.to_csv('../data/processed/one_hot_mapping.csv', index=False)
# print("\nSaved mapping to '../data/processed/one_hot_mapping.csv'")
```
