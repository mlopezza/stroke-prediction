```python
import pandas as pd
import numpy as np
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy
import PIL
import requests
import plotly as ply

```

```python
# the relative path of the stroke_positive csv dataset is saved into the variable data for ease of use
data = './data/patients_with_stroke.csv'

# The data is read and saved in a variable (stroke_positive)
stroke_positive = pd.read_csv(data)

stroke_positive = pd.DataFrame(stroke_positive)


stroke_positive
```

```python
column_names = stroke_positive.columns

print(column_names)
```

```python
# Count how many patients have each BMI
bmi_counts = stroke_positive['bmi'].value_counts().sort_index()

# Bar Plot for Count of patients by BMI
bmi_counts.plot(kind='bar', figsize=(10,6))
plt.title('Count of Patients by BMI')
plt.xlabel('BMI')
plt.ylabel('Count')
plt.show()
```

```python
# Define the bins (boundaries) of x-axis categories
bins = [0, 18.5, 25, 30, float('inf')]

# Define the labels for each bin
labels = ['Underweight', 'Healthy Weight', 'Overweight', 'Obesity']

# Create a new column for BMI category type
stroke_positive['bmi_category'] = pd.cut(
    stroke_positive['bmi'],
      bins=bins, 
      labels=labels, 
      right=False)
```

```python
# Bar Plot for Count of patients by BMI (categories)
sns.countplot(data=stroke_positive, x='bmi_category', order=labels)
plt.title('Number of Stroke Patients by BMI Type')
plt.xlabel('BMI Type')
plt.ylabel('Count')
plt.show()
```

```python
# Count how many patients have each glucose level type
glucose_counts = stroke_positive['avg_glucose_level'].value_counts().sort_index()

# Bar Plot for Count of patients by Glucose Level
glucose_counts.plot(kind='bar', figsize=(10,6))
plt.title('Number of Patients by Glucose Level')
plt.xlabel('Glucose Level')
plt.ylabel('Count')
plt.show()
```

```python
# Define the bins (boundaries) of x-axis categories
bins = [0, 100, 150, 200, float('inf')]

# Define the labels for each bin
labels = ['Low', 'Healthy', 'Pre-Diabetic', 'Diabetic']

# Create a new column for glucose categories
stroke_positive['glucose_category'] = pd.cut(
    stroke_positive['avg_glucose_level'],
    bins=bins,
    labels=labels,
    right=False
)
```

```python
# Bar Plot for Count of patients by Glucose Level (categories)

sns.countplot(data=stroke_positive, x='glucose_category', order=labels)
plt.title('Number of Patients by Glucose Level')
plt.xlabel('Glucose Category')
plt.ylabel('Count')
plt.show()
```

```python
# Count patients by age
glucose_counts = stroke_positive['age'].value_counts().sort_index()

# Bar Plot for Count of patients by age
glucose_counts.plot(kind='line', figsize=(10,6))
plt.title('Number of Patients by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
# Define the bins (boundaries) of x-axis categories
bins = [0, 14, 25, 65, float('inf')]

# Define the labels for each bin
labels = ['Children', 'Youth', 'Adult', 'Senior']

# Create a new column for age categories
stroke_positive['age_category'] = pd.cut(
    stroke_positive['age'],
    bins=bins,
    labels=labels,
    right=False
)
```

```python
# Bar Plot for Count of patients by Age (categories)

sns.countplot(data=stroke_positive, x='age_category', order=labels)
plt.title('Number of Patients by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```
